# Market Basket Analysis for Groceries dataset using Apriori and FP Growth algorithm

Loading necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import threading
import warnings
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.frequent_patterns import fpgrowth
warnings.filterwarnings("ignore")
```

Reading the Dataset

```python
grocery_data = pd.read_csv('/content/Groceries_dataset - Groceries_dataset.csv')
grocery_data
```

Checking the volume of the data

```python
grocery_data.shape
```

Making sure if any null values present in the dataset

```python
grocery_data.isna()
```

Storing the data into a list before performing Apriori/ FP Growth algorithm

```python
transactions = []
for index,row in grocery_data.iterrows():
  transactions.append(row.dropna().tolist())
transactions
```

Retrieving the values of the itemDescription and visualizing using matplotlib

```python
value = grocery_data['itemDescription'].value_counts(ascending=False)[:15]
value
```

```python
plt.figure(figsize=(25,25))
sns.barplot(x=value.index,y=value.values)
plt.show()
```

Fixing quantity for each data

```python
grocery_data['Quantity'] = 1
grocery_data
```

```python
transactions = grocery_data.groupby(['Member_number','itemDescription'])['Quantity'].sum().unstack().reset_index().set_index('Member_number')
transactions
```

  Filling values

```python
transactions = transactions.fillna(0)
transactions
```

Specifying conditions: If the value of the items less than 0 it will return 0, if it is greater it will return 1

```python
def encode(value):
  if value <= 0:
    return 0
  elif value >= 0:
    return 1

basket_analysis = transactions.applymap(encode)
basket_analysis
```

Implementing threads


```python
thread_one = threading.Event()
thread_two = threading.Event()
```

Assuming [0,0] as ap & fp

```python
execution_time = [0,0]
```

Defining seperate functions for both the algorithms

```python
def apriori_thread(min_support,min_confidence):
  stime = time.time()
  apriori_results = apriori(basket_analysis,min_support = min_support,use_colnames=True)
  execution_time[0] = time.time()-stime
  thread_one.set()
```

```python
def fp_thread(min_support,min_confidence):
  stime = time.time()
  fp_results = fpgrowth(basket_analysis,min_support = min_support,use_colnames=True)
  execution_time[1] = time.time()-stime
  thread_two.set()
```

Allowing the user to set the minimum support and confidence levels

```python
if __name__ == '__main__':
  min_support_ap = float(input("Enter minimum support for apriori algorithm : "))
  min_support_fp = float(input("Enter minimum support for fpgrowth algorithm : "))
  min_confidence_ap = float(input("Enter minimum confidence level for apriori algorithm: "))
  min_confidence_fp = float(input("Enter minimum confidence level for fpgrowth algorithm: "))
```

Assigning both the functions as target to each thread, and calling min support and confidence in args.

```python
apriorithread = threading.Thread(target=apriori_thread,args=(min_support_ap,min_confidence_ap))
fpthread = threading.Thread(target=fp_thread,args=(min_support_fp,min_confidence_fp))
```

```python
apriorithread.start()
fpthread.start()
thread_one.wait()
thread_two.wait()
apriorithread.join()
fpthread.join()
```

```python
alg = ['Apriori','Fpgrowth']
plt.bar(alg,execution_time)
plt.xlabel('Algorithm')
plt.ylabel('Execution Time')
plt.title('Apriori and FP growth Performance evaluation')
plt.show()
if execution_time[0]<execution_time[1]:
  print("Apriori is Faster, it has less execution time.")
elif execution_time[1]<execution_time[0]:
  print("FP growth is Faster, it has less execution time.")
```
