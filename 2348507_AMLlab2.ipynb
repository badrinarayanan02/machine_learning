Multiple Linear Regression

Each Sample in the dataset describes a house in Boston town. The data was drawn from the Boston Standard Metropolitan Statistical Area (SMSA) in 1970.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
data = pd.read_csv('/content/Program_1_Dataset - Program_1_Dataset.csv')
data
```

Consider the attributes Av_room, distance, Property_TaxRate Non_Retail_Acres as predictors and MValue as the target variable

```python
predictors = ['Avg_rooms','Distance','Property_TaxRate','Non_Retail_Acres']
X = data[predictors]
```

Calculating Vif for each predictor

```python
vif_data = pd.DataFrame()
vif_data['Predictors'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```
