# **Question Answering System**

Implement the Question answering system

```python
from transformers import BertTokenizer, BertForQuestionAnswering, pipeline
import torch
```

Loading the pre trained pipeline

```python
qa_pipeline = pipeline("question-answering")
```

Defining contexts and question

```python
context = "CHRIST (Deemed to be University) was born out of the educational vision of St Kuriakose Elias Chavara, an educationalist and social reformer of the nineteenth century in South India. He founded the first Catholic indigenous congregation, Carmelites of Mary Immaculate (CMI), in 1831 which administers CHRIST (Deemed to be University). “CHRIST (Deemed to be University) was established as ‘Christ College’ in 1969. It undertook path- reaking initiatives in Indian higher education with the introduction of innovative and modern curricula, insistence on academic discipline, imparting of Holistic Education and adoption of global higher education practices with the support of creative and dedicated staff."
question = "Which year CHRIST university established?"
```

Getting the answer

```python
answer = qa_pipeline(question=question, context=context)
```

```python
print("Answer:", answer["answer"])
```
