# **Exploratory data analysis**






Visualizing the distribution of price using boxplot

```python
plt.boxplot(data['price'])

plt.xlabel('Price')
plt.ylabel('Value')
plt.title('Distribution of Price')

plt.show()
```

Visualizing the relationship between bedroom and price using scatter plot

```python
plt.scatter(data['bedrooms'], data['price'])

plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Relationship between Bedrooms and Price')

plt.show()
```

Displaying the distribution of sqft living by city using boxplot

```python
plt.figure(figsize=(12, 10))

sns.boxplot(x='city', y='sqft_living', data=data)

plt.xticks(rotation=45)

plt.xlabel('City')
plt.ylabel('Sqft Living')
plt.title('Distribution of Sqft Living by City')

plt.show()
```

Distribution of bathroom across different floors

```python
mean_bathrooms_per_floor = data.groupby('floors')['bathrooms'].mean().reset_index()

sns.barplot(x='floors', y='bathrooms', data=mean_bathrooms_per_floor)

plt.xlabel('Number of Floors')
plt.ylabel('Average Number of Bathrooms')
plt.title('Average Number of Bathrooms across Floors')

plt.show()
```

Distribution of year built and year renovated

```python
year_built_counts = data['yr_built'].value_counts().sort_index()
year_renovated_counts = data['yr_renovated'].value_counts().sort_index()


plt.bar(year_built_counts.index, year_built_counts.values, alpha=0.5, label='Year Built')
plt.bar(year_renovated_counts.index, year_renovated_counts.values, alpha=0.5, label='Year Renovated')

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of Year Built and Year Renovated')

plt.legend()

plt.show()
```

Applying Bin Means and Bin Boundaries for attributes "Year Built" and "Sqft_Above"

```python
year_built_bins = [1900, 1920, 1940, 1960, 1980, 2000, 2020]

data['year_built_bin'] = pd.cut(data['yr_built'], bins=year_built_bins, include_lowest=True)

sqft_above_bins = [0, 1000, 1500, 2000, 2500, 3000, float('inf')]

data['sqft_above_bin'] = pd.cut(data['sqft_above'], bins=sqft_above_bins, include_lowest=True, right=False)

# Calculating the mean of sqft above for each year built bin

mean_sqft_above = data.groupby('year_built_bin')['sqft_above'].mean()

# Calculating the mean of year built for each sqft above bin

mean_year_built = data.groupby('sqft_above_bin')['yr_built'].mean()

print("Mean Sqft Above for Each Year Built Bin:")
print(mean_sqft_above)

print("\nMean Year Built for Each Sqft Above Bin:")
print(mean_year_built)
```
